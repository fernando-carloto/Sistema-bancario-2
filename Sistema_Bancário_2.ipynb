```python
saldo = 1000.0

def consultar_saldo():
  """Exibe o saldo atual"""
  print(f"Seu saldo atual é de R${saldo:.2f}")

def sacar():
  """Realiza a operação de saque, verificando o saldo."""
  global saldo
  while True:
    try:
      valor_saque = float(input("Digite o valor do saque: R$ "))
      break
    except ValueError:
        print("Valor inválido. Tente novamente.")

  if valor_saque <= saldo:
    saldo -= valor_saque
    print(f"Saque de R$ {valor_saque:.2f} realizado com sucesso")
    print(f"Seu saldo atual é de R$ {saldo:.2f}")
  else:
    print("Saldo insuficiente para realizar o saque.")

def depositar():
  """Realiza a operação de depósito"""
  global saldo
  while True:
    try:
      valor_deposito = float(input("Digite o valor do depósito: R$ "))
      break
    except ValueError:
      print("Valor inválido. Tente novamente.")
  saldo += valor_deposito

  print(f"Depósito de R$ {valor_deposito:.2f} realizado com sucesso")
  print(f"Seu saldo atual é de R$ {saldo:.2f}")

while True:
  print("\n--- Bem vindo ao seu banco virtual ---")
  print("1 - Consultar saldo")
  print("2 - Depositar")
  print("3 - Sacar")
  print("4 - Sair")

  while True:
    try:
        opcao_str = input("Escolha uma opção: ")
        opcao = int(opcao_str)
        if 1 <= opcao <= 4:
          break
        else:
          print("Opção inválida. Por favor, digite um número de 1 a 4.")
    except ValueError:
      print("Opção inválida. Por favor, digite um número inteiro")

  if opcao == 1:
    consultar_saldo()
  elif opcao == 2:
    depositar()
  elif opcao == 3:
    sacar()
  elif opcao == 4:
    print("Obrigado por usar nosso banco virtual. Até logo!")
    break
```

```python
import datetime

saldo = 1000.0
extrato = []


def sacar():
  """Realiza a operação de saque, verificando o saldo."""
  global saldo, extrato
  while True:
    try:
      valor_saque = float(input("Digite o valor do saque: R$ "))
      break
    except ValueError:
      print("Valor inválido. Tente novamente.")

  if valor_saque <= saldo:
    saldo -= valor_saque
    agora = datetime.datetime.now()
    extrato.append({"data_hora": agora, "tipo": "saque", "valor": valor_saque})
    print(f"Saque de R$ {valor_saque:.2f} realizado com sucesso")
    print(f"Seu saldo atual é de R$ {saldo:.2f}")
  else:
    print("Saldo insuficiente.")

def depositar():
  """Realiza a operação de depósito"""
  global saldo, extrato
  while True:
    try:
      valor_deposito = float(input("Digite o valor do depósito: R$ "))
      break
    except ValueError:
      print("Valor inválido. Tente novamente.")
  saldo += valor_deposito
  agora = datetime.datetime.now()

  extrato.append({"data_hora": agora, "tipo": "deposito", "valor": valor_deposito})
  print(f"Depósito de R$ {valor_deposito:.2f} realizado com sucesso")
  print(f"Seu saldo atual é de R$ {saldo:.2f}")

def exibir_extrato():
  """Exibe o extrato bancário"""
  global extrato
  if not extrato:
    print("Não foram realizadas movimentações.")
  else:
    print("\n--- Extrato bancário ---")
    for transacao in extrato:
      data_hora = transacao["data_hora"].strftime("%d/%m/%Y %H:%M:%S")
      tipo = transacao["tipo"]
      valor = transacao["valor"]
      print(f"{data_hora} - {tipo}: R$ {valor:.2f}")
    print(f"Saldo atual: R$ {saldo:.2f}")

while True:
  print("\n--- Bem vindo ao seu banco virtual ---")
  print("1 - Consultar saldo")
  print("2 - Depositar")
  print("3 - Sacar")
  print("4 - Exibir extrato")
  print("5 - Sair")

  while True:
    try:
        opcao_str = input("Digite a opção desejada: ")
        opcao = int(opcao_str)
        if 1 <= opcao <= 5:
          break
        else:
          print("Opção inválida. Por favor, digite um número de 1 a 5.")
    except ValueError:
      print("Opção inválida. Por favor, digite um número inteiro")

  if opcao == 1:
    consultar_saldo()
  elif opcao == 2:
    depositar()
  elif opcao == 3:
    sacar()
  elif opcao == 4:
    exibir_extrato()
  elif opcao == 5:
    print("Obrigado por usar nosso banco virtual.")
    break
```
